# Data Cleaning 

```python
df.info()
```

```python
df.drop(['Status','unnamed1'], axis = 1, inplace=True)
```

```python
df.info()
```

```python
df.isnull().sum()  # checking total null values in column 
```

```python
df.shape
```

```python
df.dropna(inplace=True)
```

```python
df.isnull().sum()  # check if null values exists
```

```python
df.info()
```

```python
# Change Data type
df['Amount']=df['Amount'].astype('int')
```

```python
df.info()
```

```python
# Rename columns

df.rename(columns={'Marital_Status' : 'Married'}, inplace=True)
```

# Exploratory Data Analysis

```python
df.head()
```

### Gender

```python
ax = sns.countplot(x = 'Gender', data = df)

for bar in ax.containers:
    ax.bar_label(bar)
```

```python
df.groupby(['Gender'],as_index= False)['Amount'].sum().sort_values(by='Amount',ascending = False)
```

```python
df.groupby(['Gender']).Amount.sum().plot(kind = 'pie')
```

```python
df.groupby(['Gender']).Amount.sum()
```

```python
sales_gen = df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending = False)
sns.barplot(x='Gender',y='Amount',data=sales_gen)
```

### Age

```python
sns.countplot(data = df, x = 'Age Group',hue='Gender')
```

```python
ax = sns.countplot(data = df, x = 'Age Group',hue='Gender')
for i in ax.containers:
    ax.bar_label(i)
```

```python
# Total amount Vs Total Group 

Sales_age = df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending =False)

sns.barplot(x= 'Age Group', y= 'Amount', data = Sales_age)
```

### State

```python
df.columns
```

```python
# total number of orders from top 10 states. 
total_Order_byStates = df.groupby(['State'],as_index=False).Orders.sum().sort_values(by='Orders',ascending = False).head(10)
```

```python
total_Order_byStates
```

```python
sns.set(rc = {'figure.figsize':(20,5)})
sns.barplot(data = total_Order_byStates, x = 'State', y ='Orders')
```

```python
# Total Amount/sales from top 10 state. 

Amount_State = df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending = False).head(10)
```

```python
sns.barplot(data = Amount_State,x='State',y='Amount')
```

### Marital_Status

```python
ax = sns.countplot(data = df, x='Married')

sns.set(rc = {'figure.figsize':(7,8)})

for i in ax.containers: 
    ax.bar_label(i)
```

- More shopping is done my married peoples.

```python
sale_state = df.groupby(['Married','Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending =False)
sns.set(rc={'figure.figsize':(7,5)})
sns.barplot(data = sale_state, x='Married',y='Amount',hue ='Gender')
```

- From Above graph we can see that most of the buyers are married women and they have high purchesing power.

### Occupation

```python
ax = sns.countplot(data=df,x='Occupation')

for i in ax.containers:
    ax.bar_label(i)

sns.set({'figure.figsize':(20,5)})
```

```python
# total Amount/Occupation 

sale_occupation = df.groupby(['Occupation'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending =False)

sns.barplot(data = sale_occupation,x='Occupation',y='Amount')
```

- From above graph we can see most of buyers are working in IT, Heathcare,Aviation

```python
df.columns
```

### Product_Category

```python
ax = sns.countplot(data=df, x='Product_Category')

for i in ax.containers:
    ax.bar_label(i)

sns.set(rc = {'figure.figsize':(50,10)})
```

```python
sale_product = df.groupby(['Product_Category'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending = False).head(10)
sns.barplot(data = sale_product, x='Product_Category',y='Amount')

sns.set({'figure.figsize' : (15,7)})
```

- From above graph we can clearly see that food products are highly saleable followed by cloths and electronics
